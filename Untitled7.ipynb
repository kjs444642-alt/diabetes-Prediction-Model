##2. 배경 및 필요성
가족 중 한분이 당뇨병에 걸려 오래동안 약을 복용하신적이 있는데 이를 보고 내외부 증상을 보고 당뇨병 여부를 판단 할 수 있는 당뇨병 예측 모델이 필요하다고 느꼈다.
##3. 목표
본 프로젝트의 목표는 개인의 건강 데이터를 기반으로 당뇨병 발병 여부를 예측할 수 있는 머신러닝 모델을 구축하는 것이다.
이를 통해 주요 건강 지표가 당뇨병에 미치는 영향을 분석하고, 예측 성능을 향상시키는 것을 목표로 한다.

특히, 실제 당뇨 환자를 놓치지 않기 위해 Recall을 중심으로 모델을 평가 및 개선하여 의료적 활용 가능성을 고려한 분석을 수행하고자 한다.
##4. 주요 기능 및 특징

본 프로젝트는 당뇨병 예측 모델 구축을 위해 다음과 같은 단계로 수행된다.

### 기능 1: 데이터 수집 및 전처리

Kaggle에서 제공하는 당뇨병 데이터셋을 활용하여 분석을 진행한다.
데이터 내 결측값 및 이상치를 확인하고, 필요에 따라 평균값 대체 및 정규화를 수행하여 모델 학습에 적합한 형태로 가공한다.

### 기능 2: 탐색적 데이터 분석 (EDA)

데이터의 전반적인 특성을 파악하기 위해 변수 간 상관관계 분석과 시각화를 수행한다.
이를 통해 당뇨병 발병에 영향을 미치는 주요 요인을 도출하고, 데이터의 분포 및 패턴을 분석한다.

### 기능 3: 머신러닝 모델 구축 및 평가

Logistic Regression과 Random Forest 알고리즘을 적용하여 당뇨병 예측 모델을 구축한다.
모델 성능은 Accuracy, Precision, Recall, ROC-AUC 지표를 활용하여 평가하며, 특히 실제 환자를 놓치지 않기 위해 Recall을 중심으로 성능을 개선한다.
또한 모델 간 성능 비교를 통해 최적의 모델을 선정한다.
## 5. 기대 효과

본 프로젝트를 통해 당뇨병 발병 여부를 사전에 예측함으로써 질병의 조기 발견 가능성을 높일 수 있다.
또한 데이터 분석을 통해 주요 건강 지표가 당뇨병에 미치는 영향을 파악하여, 예방 및 관리에 유용한 정보를 제공할 수 있을 것으로 기대된다.

더불어 머신러닝 기반의 의료 데이터 분석 과정을 수행함으로써 데이터 처리 및 모델링 능력을 향상시키고, 실제 문제 해결에 적용 가능한 데이터 사이언스 역량을 강화할 수 있다.
