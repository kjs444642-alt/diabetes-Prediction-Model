## 1. 프로젝트 진행 내용

### 1) 데이터셋 선정 및 수집

본 프로젝트에서는 Kaggle에서 제공하는 Pima Indians Diabetes Dataset을 선정하였다. 해당 데이터셋은 혈당, 혈압, BMI, 인슐린 수치, 나이 등 환자의 건강 정보를 포함하고 있으며, 이를 기반으로 당뇨병 발병 여부를 예측할 수 있도록 구성되어 있다. 의료 데이터 기반의 이진 분류 문제로 구성되어 있어 프로젝트 주제에 적합하다고 판단하였다.

### 2) 데이터 구조 및 변수 이해

데이터셋의 컬럼 및 변수 특성을 확인하고, 각 변수의 의미와 역할을 분석하였다. 주요 변수로는 혈당(Glucose), 혈압(BloodPressure), BMI, 인슐린(Insulin), 나이(Age) 등이 있으며, 최종 결과값인 Outcome은 당뇨병 발병 여부를 나타낸다. 또한 변수 간 관계를 파악하기 위해 데이터 분포와 상관관계를 확인하였다.

### 3) 기초적인 데이터 탐색 진행

데이터의 기초 통계량과 결측값 여부를 확인하였으며, 변수별 분포를 시각화하여 데이터 특성을 분석하였다. 특히 일부 변수에서 0값이 비정상적으로 존재하는 것을 확인하였고, 이를 결측값으로 처리할 필요성을 파악하였다. 또한 당뇨병 여부에 따른 변수 차이를 비교하여 주요 건강 지표와의 연관성을 분석하였다.

## 2. 현재까지의 진행 현황 및 결과

현재 Kaggle의 Pima Indians Diabetes Dataset을 활용하여 데이터 수집 및 기초 분석을 완료하였다. 데이터의 주요 변수(혈당, 혈압, BMI, 인슐린 수치, 나이 등)를 확인하고, 변수 간 관계 및 데이터 분포를 분석하였다.

또한 데이터 전처리 과정에서 일부 변수에 비정상적인 0값이 존재하는 것을 확인하였으며, 이를 결측값으로 처리하여 데이터 품질을 개선하였다. 이후 주요 변수에 대한 시각화를 수행하고, 당뇨병 여부에 영향을 미치는 주요 건강 지표를 분석하였다.

현재는 머신러닝 기반 예측 모델 구축 및 성능 평가 단계까지 진행된 상태이며, 예측 결과를 단순 이진 분류가 아닌 위험도 등급 형태로 확장하는 방향으로 프로젝트를 발전시키고 있다. 기존 프로젝트에서는 결과를 단순히 당뇨병 여부로만 분류하였으나, 본 프로젝트에서는 예측 확률을 기반으로 저위험, 중위험, 고위험으로 구분하는 위험도 등급 방식을 추가하여 보다 직관적이고 활용 가능한 결과를 제공하고자 한다.
